```python
# I begin with creating the employee superclass 
class Employee:
    # I then begin to set the constructor method to run when an employee object is created 
    def __init__(self, name, number):
        # I then create a private attribute for employee name 
        self.__name = name 
        # I then do the same for employee number 
        self.__number = number 
    # I then work on returning the values stored within name and number 
    def get_name(self):
        # This is the function I used to return the value stored in __name 
        return self.__name 
    # I then do the same for employee number starting with the def function 
    def get_number(self):
        # I then use return as I did above, however for __number this time 
        return self.__number
    # I then move on to the mutator methods 
    def set_name(self, name):
        # I then create a function allowing employee names to be updated 
        self.__name=name
    # I then do the same for number 
    def set_number(self, number):
        # I then create a funciton that updates the employee number 
        self.__number=number
# I then move on the the subclass 
class ProductionWorker(Employee):
    # I  then get this subclass to inherit all attributes from the previous superclass 
    def __init__ (self, name, number, shift, hourly_pay_rate ):
        # I then call the employee constructor 
        super().__init__(name, number)
        # I then create an private attribute for shift 
        self.__shift = shift
        # I then do the same for the hourly_pay_rate
        self.__hourly_pay_rate = hourly_pay_rate
    # I then use the accesor methods to ruturn the shift number 
    def get_shift(self):
        # this function returns the shift number 
        return self.__shift 
    # I then begin the same for the hourly_pay_rate 
    def get_hourly_pay_rate(self):
        # I the use the same return function from above to return the hourly_pay_rate
        return self.__hourly_pay_rate
    # I then use the mutator methods to edit shift numbers 
    def set_shift(self, shift):
        # this function updates the number stored 
        self.__shift=shift 
    # I then do the same for the hourly_pay_rate
    def set_hourly_pay_rate(self, hourly_pay_rate):
        # allows the rate to be updated 
        self.__hourly_pay_rate = hourly_pay_rate
# I then move onto the main function that creates and uses a production worker objec
def main():
# I print user to input workers info
    print("Enter production worker Info:")
    # seperation for neatness 
    print("------------------------------") 
    # prompt user for employee name 
    name = input("Enter employee name:")
    # same for number 
    number = input("Enter employee number:")
    # I then ask for shift while converting to float for storage 
    shift = int(input("Enter shift number (1=day, 2=night):"))
    # I then ask for the hourly_pay_rate
    hourly_pay_rate=float(input("Enter hourly pay rate:"))
    # I then create a productionworker object using all four attributes 
    worker = ProductionWorker(name, number, shift, hourly_pay_rate)
    # I then display info using the accessor method
    print("\nEmployee Information")
    # lines for seperation 
    print("------------------------------------------")
    # retrieve name 
    print("Name:", worker.get_name())
    #then number 
    print("Employee Number:", worker.get_number())
    # then shift number 
    print("Shift:", worker.get_shift())
    # then hourly rate 
    print("Hourly pay rate:$", worker.get_hourly_pay_rate(), '.2f', sep = "")
    # calling main to run program
main() 
```
