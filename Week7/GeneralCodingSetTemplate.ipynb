Total Sales

Design a program that asks the user to enter a storeâ€™s sales for each day of the week. The amounts should be stored in a list. Use a loop to calculate the total sales for the week and display the result.

#### High-Level Overview

*In two to three sentences at minimum, walk through your general approach to the problem and identify any particular issues you had.*


```python
# I beging with creating a list to hold seven days of sales 
sales=[0]*7
# I then create the loop that gathers the sales for each days 
for i in range(7):
    # Here is where I gather user input for the sales of the seven days 
    sales[i]= float(input(f"Enter the sales for day {i +1}:"))
    # I then create the accumulator variable that will store the total 
    total = 0 
    # I then create the loop that will calculate the sum of the sales 
for amount in sales: 
    # I had each days sales to the total 
    total+=amount 
# I then print the total sales of the week 
print(f"The total sales for the week is: {total}")
```
